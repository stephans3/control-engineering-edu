# Basic Systems and Block Diagrams

## First Order System

The dynamics of output $y(t)$ is described by the differential equation

$$
T ~ \dot{y}(t) + y(t) = K ~ u(t)
$$

with the initial value $x(0) = x_{0}$, time constant $T > 0$, gain $K > 0$ and input $ u \in \mathcal{PC}(\mathbb{R},\mathbb{R}) $ (piecewise-continuous function).  The differential equation is transformed to 

$$
\left[sT + 1\right] Y(s) = K ~ U(s)
$$

and reshaped with frequency $ \omega = \frac{1}{T} $ to

$$
G(s) = \frac{Y(s)}{U(s)} = \frac{K}{T~s + 1} = \frac{K \omega}{s + \omega} \text{.}
$$

The relationship between input $U(s)$ and output $Y(s)$ of a dynamical system is noted as the transfer function. Furthermore, transfer function $G$ can be noted either in a polynomial / time-constant form as

$$
G(s) = \frac{K}{T~s + 1}
$$

which is useful for the representation in source code (e.g. MATLAB), or in a pole-zero form 

$$
G(s) = \frac{K \omega}{s + \omega} \text{.}
$$

which is useful to find the poles, here: $s = -\omega$.

## Second Order System - Series Connection

Second order systems are either derived from second order differential equations as in the previous sections or they are modeled directly as the multiplication of first order transfer functions. 

Consider two transfer functions

$$
G_{1}(s) = \frac{X(s)}{U(s)} = \frac{1}{T_{1}~s + 1}
$$

and

$$
G_{2}(s) = \frac{Y(s)}{X(s)} = \frac{1}{T_{2}~s + 1}
$$

with time constants $T_{1}, T_{2} > 0$. Here, the output of $G_{1}$ and the input of $G_{2}$ are identical: state $X(s)$. Therefore, both first order systems can be connected and the multiplication of both transfer functions results in the second order system

$$
G_{sc}(s) = G_{1}(s) ~ G_{2}(s) = \frac{Y(s)}{U(s)}  = \frac{1}{(T_{1}~s + 1) ~ (T_{2}~s + 1)}
$$

which can be formulated in the polynomial form as

$$
G_{sc}(s) = \frac{1}{T_{1}~T_{2}~s^2 + (T_{1}+T_{2}) ~ s + 1} \text{.}
$$

This multiplication corresponds to a **series connection** of two blocks of first order systems. 

State $X(s)$ does not appear anymore directly in system $G_{sc}$ but it is still included - see the figure below.

### Simulation

The series connection is simulated with $T_{1} = 0.1$ seconds, $T_{2} = 10$ seconds and a final simulation time $T = 50$ seconds.The input is given as the step function.

```julia
using ControlSystems

T1 = 0.1; # Time constants
T2 = 10;
Tf = 50; # Final simulation time

G1 = tf(1, [T1, 1]);
G2 = tf(1, [T2, 1]);
Gsc = G1 * G2 # Series connection
```

```julia
stepplot(Gsc, Tf, label="y(t)")
```

The output of each subsystem ($x(t)$ and $y(t)$) tends to $1.0$ for $t \rightarrow \infty$. Thus, the multiplication leads to an output of $1.0$ for $t \rightarrow \infty$, too.

## Parallel connection


Two or more blocks of arbitrary dynamical systems can be parallel connected via summation. The two blocks

$$
G_{1}(s) = \frac{Y_{1}(s)}{U(s)} = \frac{1}{T_{1} ~ s + 1}
$$

and

$$
G_{2}(s)= \frac{Y_{2}(s)}{U(s)}  = \frac{1}{T_{2} ~ s + 1}
$$

are parallel connected and one yields for the entire system's output

$$
Y(s) = Y_{1}(s) + Y_{2}(s) = \left[G_{1}(s) + G_{2}(s)\right] ~ U(s) = G_{pc}(s) ~ U(s)
$$

with

$$
G_{pc}(s) = \frac{1}{T_{1} ~ s + 1} + \frac{1}{T_{2} ~ s + 1} = \frac{(T_{1} + T_{2}) ~ s + 1}{(T_{1} ~ s + 1)~(T_{2} ~ s + 1)} \text{.}
$$

## Feedback


Control systems contain feedback loops to stabilize systems or processes and to track input references. Feedback loops for the controller design will be discussed in a subsequent section. Here, the elementary structure of feedback loops will be explained. 

Consider a feedforward branch with system $G_{1}$, input $E(s)$ and output 

$$
Y(s) = G_{1}(s) ~ E(s) 
$$

and a feedback branch with system $G_{2}$, input $Y(s)$ and output

$$
Y_{2}(s) = G_{2}(s) ~ Y(s) \text{.}
$$

The feedforward and feedback branches are connected via 

$$
E(s) = R(s) - Y_{2}(s)
$$

with reference $R(s)$. The output of the entire system is calculated with

$$
Y(s) = G_{1}(s) ~ E(s) = G_{1}(s) ~ \left[R(s) - Y_{2}(s)\right] = G_{1}(s) ~ \left[R(s) - G_{2}(s) ~ Y(s)\right] 
$$

and further reshaped to yield 

$$
Y(s) = \frac{G_{1}(s)}{1 + G_{1}(s)~G_{2}(s)} ~ R(s) \text{.}
$$

The transfer function of this feedback system is

$$
G_{fb}(s) = \frac{Y(s)}{R(s)} = \frac{G_{1}(s)}{1 + G_{1}(s)~G_{2}(s)} \text{.}
$$
